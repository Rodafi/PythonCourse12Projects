# Parte 2: Melhores e piores filmes.

```python
from IPython.display import HTML
```

```python
data_best=data[["poster_path","title","budget_musd","revenue_musd","vote_count","vote_average","popularity"]].copy()
data_best.head(5)
```

```python
data_best["profit_musd"]=data.revenue_musd.sub(data.budget_musd)
#data_best["profit_musd"]=data.revenue_musd - data.budget_musd
data_best["return"]=data.revenue_musd.div(data.budget_musd)
data_best.head(5)
```

```python
print(list(data_best.columns))
```

```python
data_best.columns=["","Title","Budget","Revenue","Votes","Average Rating","Popularity","Profit","ROI"]
```

```python
data_best.set_index("Title", inplace=True)
```

```python
data_best.head()
data_best[data_best.index=="Cinderella"]
```

```python
subset=data_best.iloc[:5,:2]
subset
```

```python
data_best.iloc[0,0]
```

```python
pd.options.display.max_colwidth = 200
HTML(subset.to_html(escape=False))
```

```python
data_best.sort_values(by = 'Average Rating', ascending=False)
```

```python
data_best.sort_values(by = 'ROI', ascending=False)
```

```python
data_best[data_best.Budget>=5].sort_values(by = 'ROI', ascending=False)
```

```python
data_best.Budget.fillna(0,inplace=True)
data_best.Votes.fillna(0,inplace=True)
```

```python
data_best.info()
```

```python
data_best.loc[(data_best.Votes>=5)].head(5)
```

```python
def best_worst(n,by,ascending=False,min_bud=0,min_votes=0):
    df2=data_best.loc[(data_best.Budget>=min_bud)&(data_best.Votes>=min_votes),["",by]].sort_values(by=by, ascending=ascending).head(n).copy()
    
    return HTML(df2.to_html(escape=False))
    
```

```python
best_worst(n=5,by="Revenue")
```

```python
best_worst(5,"Budget")
```

```python
best_worst(5,"Profit")
```

```python
best_worst(5,"Profit", ascending=True)
```

```python
best_worst(5,"ROI",ascending=True, min_bud=50)
```

```python
print(data.production_companies.describe())
print(data.production_companies.isna().sum())
```

```python
filter= (data.production_companies.str.contains("Pixar").fillna(False)) & (data.release_date.between("2010-01-01","2015-12-31"))
print(filter.describe())
result =data.loc[filter,["title","poster_path","revenue_musd","release_date"]].sort_values(by="revenue_musd", ascending =False).set_index("title")
HTML(result.to_html(escape=False))
```

```python
filter = (data.genres.str.contains("Action") | data.genres.str.contains("Thriller")) & (data.original_language=="en")
print(filter.describe())
result =data.loc[filter,["title","poster_path","genres","vote_average","vote_count","release_date"]].sort_values(by="release_date", ascending =False).set_index("title").head(20)
HTML(result.to_html(escape=False))
```

```python
from wordcloud import WordCloud
```

```python
teste=data.head(5).copy()
print(teste.tagline)
print(teste.tagline.dropna())
```

```python
title=data.title.dropna();
overview=data.overview.dropna();
tagline=data.tagline.dropna();
```

```python
title_joined=' '.join(title)
overview_joined=' '.join(overview)
tagline_joined=' '.join(tagline)
print(tagline_joined)
```

```python
title_wordcloud=WordCloud(background_color='White', height=2000, width=4000, max_words=200).generate(title_joined)
plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud,interpolation='bilinear')
plt.axis=('off')
plt.show()
```

```python
title_wordcloud=WordCloud(background_color='White', height=2000, width=4000, max_words=200).generate(tagline_joined)
plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud,interpolation='bilinear')
plt.axis=('off')
plt.show()
```

```python
title_wordcloud=WordCloud(background_color='White', height=2000, width=4000, max_words=200).generate(overview_joined)
plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud,interpolation='bilinear')
plt.axis=('off')
plt.show()
```
