```
movie_id=[0,299534,19995,140607,299536,597,135397,420818,
          24428,168259,99861,254054,12445,181808,330457,
          351286,109445,321612,260513]
api_key='47846eb512c093c80c726b785489c7df'
basic_url='https://api.themoviedb.org/3/movie/{}?api_key={}'
json_list=[]
for movie in movie_id:
  print(movie)
  url=basic_url.format(movie,api_key)
  r=requests.get(url)
  if r.status_code !=200:
    print("Erro")
    continue
  else:
    data=r.json()
    json_list.append(data)
    print("Sucesso")
df=pd.DataFrame(json_list)
df
```
